**Example of historical posts extraction**

```python
username = df["username"].dropna().unique()[0]
print(f"Testing with username: {username}")
```

  **Some Definitions**

-  Handle:
  username on Bluesky.

-  DID (Decentralized Identifier):
  A unique, permanent ID assigned to each Bluesky user.  
  Unlike handles, DIDs never change and are used internally to fetch data.

-  Feed:  
  A collection of posts.


```python
try:
    #Resolve user DID
    resolved = client.com.atproto.identity.resolve_handle({"handle": username})
    did = resolved.did

    #Fetch historical posts
    posts = []
    resp = client.app.bsky.feed.get_author_feed({"actor": did, "limit": 50})
    posts.extend(resp.feed)
    cursor = resp.cursor

    while cursor:
        resp = client.app.bsky.feed.get_author_feed({"actor": did, "limit": 50, "cursor": cursor})
        posts.extend(resp.feed)
        cursor = resp.cursor

    # Save JSON
    json_path = os.path.join(folder_path, f"{username.replace('.', '_')}_test.json")
    with open(json_path, "w", encoding="utf-8") as f:
        json.dump([item.model_dump() for item in posts], f, ensure_ascii=False, indent=2)


    # Save CSV
    csv_records = [
    {
        "post_uri": item.post.uri,
        "text": getattr(item.post.record, "text", ""),
        "created_at": getattr(item.post.record, "created_at", ""),
        "author": username,
    }
    for item in posts
]

    csv_path = os.path.join(folder_path, f"{username.replace('.', '_')}_test.csv")
    pd.DataFrame(csv_records).to_csv(csv_path, index=False, encoding="utf-8")

    print("Test fetch & save complete!")

except Exception as e:
    print(f"Unexpected error: {e}")
```

# **Extract historical posts for the first list**

```python
client = Client()
```

```python
from getpass import getpass
username_bsky = input("Enter your Bluesky username (handle or email): ")
password_bsky = getpass("Enter your Bluesky password: ")
client.login(username_bsky, password_bsky)
```

```python
df = pd.read_excel(input_file_path)
usernames = df["username"].dropna().unique().tolist()
skipped_users = []
```

```python
len(usernames)
```

Bluesky allows ~3000 requests every 5 minutes per account.
So we added a short pause after each request

```python
for username in usernames:
    print(f"\n Processing: {username}")
    json_path = os.path.join(folder_path, "users_posts_json", f"{username.replace('.', '_')}.json")
    # csv_path = os.path.join(folder_path, "users_posts_csv", f"{username.replace('.', '_')}.csv")

    if os.path.exists(json_path):
        print(f" Skipping {username} ‚Äî userposts already collected")
        continue
    try:
        # Resolve DID
        resolved = client.com.atproto.identity.resolve_handle({"handle": username})
        did = resolved.did
        time.sleep(1.5)

        # Fetch posts

        posts = []
        resp = client.app.bsky.feed.get_author_feed({"actor": did, "limit": 50})
        posts.extend(resp.feed)
        cursor = resp.cursor
        time.sleep(1.5)
        while cursor:
            resp = client.app.bsky.feed.get_author_feed({"actor": did, "limit": 50, "cursor": cursor})
            posts.extend(resp.feed)
            cursor = resp.cursor
            time.sleep(1.5)

        # Save JSON
        with open(json_path, "w", encoding="utf-8") as f:
            json.dump([item.model_dump() for item in posts], f, ensure_ascii=False, indent=2)

        print(f" Saved {len(posts)} posts for {username}")

    except Exception as e:
        print(f" Skipped {username} due to error: {e}")
        skipped_users.append({"username": username, "error": str(e)})
        time.sleep(2)
        continue
if skipped_users:
  skipped_path = os.path.join(folder_path, "skipped_users.json")
  with open(skipped_path, "w", encoding="utf-8") as f:
      json.dump(skipped_users, f, ensure_ascii=False, indent=2)

print("\n Done! All usernames processed.")
```

```python
# Check already processed users
processed_files = set(os.listdir(os.path.join(folder_path, "users_posts_json")))
processed_usernames = {
    filename.replace(".json", "").replace("_", ".") for filename in processed_files if filename.endswith(".json")
}

# Compute remaining usernames
remaining_usernames = [u for u in usernames if u not in processed_usernames]

print(f"\nRemaining users to collect posts for: {len(remaining_usernames)}")
# print("Usernames:")
# for u in remaining_usernames:
#     print(u)
```

```python
import os
import json


data_dir = "/content/drive/MyDrive/Nour/users_posts_json"

# Initialize counters
user_count = 0
total_posts = 0

# Loop through files in the directory
for filename in os.listdir(data_dir):
    if filename.endswith('.json'):
        user_count += 1
        file_path = os.path.join(data_dir, filename)
        with open(file_path, 'r') as f:
            try:
                data = json.load(f)
                total_posts += len(data)
            except json.JSONDecodeError:
                print(f"Could not read {filename} (not valid JSON)")

print(f"Total users collected: {user_count}")
print(f"Total posts collected: {total_posts}")
```
